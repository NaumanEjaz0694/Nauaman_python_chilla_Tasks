```python
# Customizing plots in Python
```

```python
import seaborn as sns
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
kashti=sns.load_dataset('titanic')
kashti
```

```python
kashti.head()
```

```python
sns.boxplot(x='survived',y='age',showmeans=True
               ,meanprops=
               {'marker':'*','markersize':'12','markeredgecolor':'red'},data=kashti)
plt.xlabel('How many Survived',size=14),
plt.ylabel('Age in Years',size=14),
plt.title('Box plot of kitnay dobay aur kitnay bach gaye',size=16)
plt.show()

```

### To make the title and the legends bold just write weight="bold"

```python
sns.boxplot(x='survived',y='age',showmeans=True
               ,meanprops=
               {'marker':'*','markersize':'12','markeredgecolor':'red'},data=kashti)
plt.xlabel('How many Survived',size=14,weight='bold'),
plt.ylabel('Age in Years',size=14,weight='bold'),
plt.title('Box plot of kitnay dobay aur kitnay bach gaye',size=16,weight='bold')
plt.show()
```
